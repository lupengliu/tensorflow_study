```python
import tensorflow as tf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.linspace(-0.5,0.5,500)[:,np.newaxis]
noise = np.random.normal(0,0.02,x_data.shape)
y_data = np.square(x_data) + noise
```

```python
x = tf.placeholder(tf.float32, shape=[None, 1])
y = tf.placeholder(tf.float32, shape=[None, 1])

#定义神经网络中间层
w_1 = tf.Variable(tf.random_normal([1,10]))
b_1 = tf.Variable(tf.zeros([1,10]))

w_2 = tf.Variable(tf.random_normal([10,1]))
b_2 = tf.Variable(tf.zeros([1,1]))

h_1 = tf.add(tf.matmul(x,w_1), b_1)
L_1 = tf.nn.tanh(h_1)

h_2 = tf.add(tf.matmul(L_1,w_2), b_2)

loss = tf.reduce_mean(tf.square(h_2-y))

optimizer = tf.train.GradientDescentOptimizer(1e-2)

train = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()


with tf.Session() as sess:
    sess.run(init)
    for step in range(100001):
        sess.run(train, feed_dict={x:x_data, y:y_data})
        if step % 20 == 0:
            print(sess.run(loss, feed_dict={x:x_data, y:y_data}))
            
    prediction = sess.run(h_2, feed_dict={x:x_data, y:y_data})
    plt.figure()
    plt.scatter(x_data, y_data)
    plt.plot(x_data, prediction, 'r-',lw=5)
```
